```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
data = r'C:\Users\abhin\Desktop\UsefulBooks\Datasets\world-happiness\2019.csv'
```

```python
import pandas as pd
df = pd.read_csv(data)
```

```python
df.columns
```

```python
model.fit(df[['GDP per capita']],df[['Healthy life expectancy']])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df[['GDP per capita']] , df[['Healthy life expectancy'] ], color = "red")
plt.show
```

```python
y = model.predict(df[['GDP per capita']])
```

```python
plt.scatter(df[['GDP per capita']] , df[['Healthy life expectancy'] ], color = "red")
plt.plot(df[['GDP per capita']] , y , color = "green")
plt.xlabel("GDP per Capita")
plt.ylabel('Healthy life expectancy')
plt.show
```
